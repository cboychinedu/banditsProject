```python
# Importing the necessary modules 
from ultralytics import YOLO
import torch 
```

```python
# Optional: Check if a GPU is available (recommended for faster training)
if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
    device = torch.device("cpu")
    print("GPU not found. Training on CPU (will be slower).")

# 1. Load a pretrained model (YOLOv8 small)
# The .pt file will be automatically downloaded if it doesn't exist locally.
model = YOLO('yolov8s.pt') 
```

```python
# 2. Train the model
# The training process will print real-time updates directly into the Notebook cell's output.
results = model.train(
    data='data.yaml',   # Path to your YAML file
    epochs=50,          # Number of training epochs
    imgsz=640,          # Image size
    batch=16,           # Number of images per batch 
    name='customYOLOv8Run',
    # Optional: Pin the training to the GPU if available (otherwise uses all available devices)
    # device=0 # Use this if you have multiple GPUs and want to specify one
)
```
